```python
# merge products, aisles and departments df for readability
merge_products = pd.merge(products, aisles, on="aisle_id", how="left")
merge_products = pd.merge(merge_products, departments, on="department_id", how="left")
merge_products = merge_products[["product_id", "product_name", "aisle_id", "aisle", "department_id", "department"]]
merge_products.head(15)
```

```python
# add product names to order_products_prior
merge_o_p_prior = pd.merge(orders_prior, products[['product_id','product_name']], on='product_id', how="left")
merge_o_p_prior.head()
```

```python
# which products are purchased often? --> explore if there is a trend in popular products
pop_products = merge_o_p_prior.groupby(['product_name'])
pop_products_df = pd.DataFrame(pop_products['reordered'].sum())
pop_products_df = pop_products_df.sort_values('reordered', ascending=False).nlargest(10, 'reordered')
```

```python
pop_products_df.reset_index(inplace=True)
pop_products_df.head(10)
```

```python
sns.set_style("whitegrid")
barplt = sns.barplot(x='product_name',y="reordered", data=pop_products_df, color="mediumseagreen")
plt.xticks(rotation='75', fontsize=8, ha='right')
barplt.set_title("Bestselling Products", fontsize=16)
barplt.set_xlabel("Product Name", fontsize=12)
barplt.set_ylabel("# of Reorders", fontsize=12)

plt.gcf()
plt.tight_layout()
plt.savefig("bestselling_products.png", dpi=500)
```

```python
# what times are products frequently purchased? --> explore when are customers most likely to order
purchase_times_df = pd.merge(merge_o_p_prior, orders[['order_id', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']], on='order_id', how='left')
purchase_times_df.head(10)
```

```python
grouped_purchase_times = purchase_times_df.groupby(['order_dow', 'order_hour_of_day'], as_index=False).count()
```

```python
counts_purchase_times = pd.DataFrame(grouped_purchase_times)
counts_purchase_times.head()
```

```python
sns.set_style("whitegrid")
ptg = sns.FacetGrid(counts_purchase_times, col="order_dow", col_wrap=3, size=5, aspect=1, margin_titles=True)
ptg.map(sns.barplot, "order_hour_of_day", "order_id", color="mediumseagreen")

plt.subplots_adjust(top=.9)
ptg.fig.suptitle('Busiest Times (Amount of Orders per Hour across Week)', fontsize=16)

axes = ptg.axes.flatten()
axes[0].set_title("Sunday")
axes[1].set_title("Monday")
axes[2].set_title("Tuesday")
axes[3].set_title("Wednesday")
axes[4].set_title("Thursday")
axes[5].set_title("Friday")
axes[6].set_title("Saturday")

ptg.set_axis_labels('Hour of Day', 'Amount of Orders')
    
plt.savefig("busiest times.png", dpi=500)
```

```python
grouped_days_since = purchase_times_df.groupby(['days_since_prior_order'], as_index=False).count()
counts_days_since = pd.DataFrame(grouped_days_since)
counts_days_since.head(10)
```

```python
sns.set_style("whitegrid")
dsg = sns.barplot(x='days_since_prior_order',y="order_id", data=counts_days_since, color="mediumseagreen")

plt.xticks(rotation='75', fontsize=8, ha='center')
dsg.set_title("Purchasing Distribution", fontsize=16)
dsg.set_xlabel("Days Since Prior Order", fontsize=12)
dsg.set_ylabel("# of Orders", fontsize=12)

plt.gcf()
plt.tight_layout()
plt.savefig("Purchasing Distribution", dpi=500)
```

```python
# which products are added to the cart first? how many of these are reorders? --> could possibly help with training the model
added_first = purchase_times_df.loc[purchase_times_df["add_to_cart_order"] == 1, ["order_id", "product_name", "add_to_cart_order", "reordered"]]
added_first.head()
```

```python
added_first.reset_index(inplace=True, drop=True)
added_first.head()
```

```python
first_prod = added_first["product_name"].value_counts()
first_prod = first_prod.rename_axis('product_name').reset_index(name='counts')
first_prod = first_prod.sort_values('counts', ascending=False).nlargest(10, 'counts')
first_prod.head()
```

```python
sns.set_style("whitegrid")
barplt = sns.barplot(x='product_name',y="counts", data=first_prod, color="mediumseagreen")
plt.xticks(rotation='75', fontsize=8, ha='right')
barplt.set_title("First in Cart", fontsize=16)
barplt.set_xlabel("Product Name", fontsize=12)
barplt.set_ylabel("# of Orders", fontsize=12)

plt.gcf()
plt.tight_layout()
plt.savefig("First in Cart", dpi=500)
```

```python
added_first_reorder = added_first.loc[added_first["reordered"] == 1, ["order_id", "product_name", "add_to_cart_order", "reordered"]]
added_first.reset_index(inplace=True, drop=True)
added_first_reorder.head()
```

```python
first_prod_reorder = added_first_reorder["product_name"].value_counts()
first_prod_reorder = first_prod_reorder.rename_axis('product_name').reset_index(name='counts')
first_prod_reorder = first_prod_reorder.sort_values('counts', ascending=False).nlargest(10, 'counts')
first_prod_reorder.head(10)
```

```python
sns.set_style("whitegrid")
barplt = sns.barplot(x='product_name',y="counts", data=first_prod_reorder, color="mediumseagreen")
plt.xticks(rotation='75', fontsize=8, ha='right')
barplt.set_title("First in Cart - Reorders", fontsize=16)
barplt.set_xlabel("Product Name", fontsize=12)
barplt.set_ylabel("# of Orders", fontsize=12)
barplt.set_ylim([0,117000])

plt.gcf()
plt.tight_layout()
plt.savefig("First in Cart - Reorders", dpi=500)
```

```python
# which products are ordered once and never again?
first_no_reorder = added_first.loc[added_first["reordered"] == 0, ["order_id", "product_name", "add_to_cart_order", "reordered"]]
first_no_reorder.reset_index(inplace=True, drop=True)
first_no_reorder.head()
```

```python
sns.set_style("whitegrid")
barplt = sns.barplot(x='product_name',y="counts", data=first_no_reorder, color="mediumseagreen")
plt.xticks(rotation='75', fontsize=8, ha='right')
barplt.set_title("First in Cart - No Reorders", fontsize=16)
barplt.set_xlabel("Product Name", fontsize=12)
barplt.set_ylabel("# of Orders", fontsize=12)
# barplt.set_ylim([0,117000])

plt.gcf()
plt.tight_layout()
plt.savefig("First in Cart - No Reorders", dpi=500)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
### splitting prior orders data into train and test batches. 
### Applied Random Forest Regressor and fit to train data. 
### Next step was to test on test data and validate with orders train data. 
orders_prior = pd.read_csv("order_products__prior.csv")
orders_prior = pd.merge(orders_prior, orders[['order_id', 'order_dow', 'order_hour_of_day']], on='order_id', how='left')
orders_prior.head()
```

```python
orders_prior = orders_prior[['order_id', 'order_dow', 'order_hour_of_day', 'add_to_cart_order', 'reordered', 'product_id']]
orders_prior.head()
```

```python
target = orders_prior["product_id"]
```

```python
data = orders_prior.drop("product_id", axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = RandomForestRegressor(n_estimators=10, verbose=3)
clf = clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
